```python
from ast import literal_eval

def getPropListByJsonList(x, prop):
    if pd.isnull(x):
        return []
    jsonList = literal_eval(str(x))
    propList = []
    for val in jsonList:
        propList.append(val.get(prop))
    return propList
```

```python
movies['genres_names'] = movies.genres.apply(lambda x: getPropListByJsonList(x, 'name'))
movies['production_countries_list'] = movies.production_countries.apply(lambda x: getPropListByJsonList(x, 'name'))
movies['production_companies_list'] = movies.production_companies.apply(lambda x: getPropListByJsonList(x, 'name'))
movies['spoken_languages_list'] = movies.spoken_languages.apply(lambda x: getPropListByJsonList(x, 'iso_639_1'))
```

```python
movies = movies.drop(['belongs_to_collection', 'genres', 'production_countries', 'production_companies', 'spoken_languages'],1)
```

```python
movies.head()
```

```python
movies.to_csv('data/preprocessed_movies.csv')
```

```python
movies_exploded_by_genre = movies.explode("genres_names")
movies_exploded_by_genre.to_csv('data/movies_exploded_genres.csv')

movies_exploded_by_production_countries = movies.explode("production_countries_list")
movies_exploded_by_production_countries.to_csv('data/movies_exploded_production_countries.csv')

movies_exploded_by_production_companies = movies.explode("production_companies_list")
movies_exploded_by_production_companies.to_csv('data/movies_exploded_production_companies.csv')
```

```python
genres = movies_exploded_by_genre.groupby("genres_names")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
genres.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Genres")
plt.ylabel("Popularity")
plt.show()
```

```python
plt.figure(figsize=(15,10))
genres.mean().sort_values(by="popularity",ascending=False)["popularity"].plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Genres")
plt.ylabel("Popularity")
plt.show()
```

```python
plt.figure(figsize=(15,10))
genres.mean().sort_values(by="vote_average",ascending=False)["vote_average"].plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Genres")
plt.ylabel("Vote average")
plt.show()
```
